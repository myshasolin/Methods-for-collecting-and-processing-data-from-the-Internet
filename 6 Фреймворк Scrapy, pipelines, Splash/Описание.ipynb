Леруа сразу заблочил меня и поиграться с собой не дал, жаль. Куки не помогли. Поэтому я парсил Кастораму.

Собираю информацию по ударным дрелям (нужна была недавно, хорошую найти оказалось не так и просто), собираю:

- **‘good_name’** – название
- **‘price’** – цена (тип int)
- **‘link’** – ссылка на товар
- **‘images’** – массив изображений с ключами: url, path, checksum и status
- **‘product_characteristic’** – сбор индивидуальных характеристик товаров со страницы

За счёт определения пути в методе file_path фотографии скачиваются складываются в отдельные именные папки, которые формируются по названию товаров. Сложил все собранные фотки на гугл диск. 

А вся остальная собранная информация собирается в БД db_hardware_store в коллекцию Castorama_ru, сейчас в ней 168 документов. JSON-файл с данными здесь.
